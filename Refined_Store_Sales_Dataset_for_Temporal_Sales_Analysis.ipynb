```python
import pandas as pd
```

```python
df = pd.read_csv( "/content/drive/MyDrive/ELEVATE LABS/store_sales_data (2).csv")
df
```

```python
df.isna().sum()
```

```python
# Step 1: Remove rows with any null values
df.dropna(inplace=True)

# Step 2: Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Step 3: Remove any rows where 'Order Date' couldn't be parsed
df = df.dropna(subset=['Order Date'])

# Step 4: Create new 'Month-Year' column
df['Month-Year'] = df['Order Date'].dt.strftime('%b-%Y')

# Step 5: Preview the result
df
```

```python
excel_path = "/content/cleaned_store_sales_data.xlsx"
df.to_excel(excel_path, index=False)
```
